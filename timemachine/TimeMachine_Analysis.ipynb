```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()
llm = ChatGroq(model="llama-3.1-8b-instant")
```

```python
with open("knowledge_snapshot_2025-01-15.txt", "r") as f:
    knowledge = f.read()
print(knowledge[:200])
```

```python
extraction_prompt = f"""
You are a Knowledge Extraction Engine.

Your task is to analyze a personal knowledge snapshot and extract all knowledge into a structured JSON format.

Rules:

1. Extract the date.
2. Extract the person's name.
3. Identify each knowledge category.
4. Extract every factual statement.
5. Distinguish between:

   * Known Skills
   * Learning In Progress
   * Goals
   * Projects
   * Confidence Areas
   * Knowledge Gaps
6. Do not invent information.
7. Preserve the original meaning.
8. Generate concise factual records suitable for a vector database.
9. For each fact assign:

   * date
   * category
   * fact
   * confidence_level
   * status

Status values:

* known
* learning
* goal
* project
* gap

Output JSON only.

Input:

Date: 2025-01-15
Person: Ahmad Shajahan

Knowledge Snapshot:

Programming:

* Learned Python fundamentals.
* Understood variables, loops, functions, and classes.
* Built small command-line applications.
* Started learning NumPy and Pandas.

Artificial Intelligence:

* Basic understanding of Machine Learning.
* Learned about supervised and unsupervised learning.
* Familiar with neural networks conceptually.
* Has not yet built a complete AI project.

Web Development:

* Learned HTML, CSS, and JavaScript.
* Built simple websites.
* Started learning Django.

Projects:

* Created a coding school website named Mrcode.
* Started working on AI-related educational content.

Current Goals:

* Learn advanced Machine Learning.
* Build AI-powered educational tools.
* Improve full-stack development skills.

Confidence Areas:

* Python programming
* Basic web development

Areas Still Learning:

* Deep Learning
* Large Language Models (LLMs)
* RAG systems
* Multi-agent AI systems
knowledge_snapshot:
{knowledge}
"""

extraction_response = llm.invoke(extraction_prompt)
extracted_value = extraction_response.text

print("===STAGE 1 : EXTRACTED VALUE ===")
print(extracted_value)
```

```python

prompt = f"""
You are an expert Knowledge Extraction Engine.

Your task is to analyze the provided text and convert it into structured memories for a Time Machine Knowledge System.

Instructions:

1. Extract all important facts.

2. Extract date if present.

3. Extract person name if present.

4. Identify:

   * skills
   * technologies
   * projects
   * achievements
   * goals
   * interests
   * learning activities
   * knowledge gaps

5. One memory record = one fact.

6. Do not invent information.

7. Preserve original meaning.

8. Return ONLY valid JSON.

9. No markdown.

10. No explanations.

Status values:

* known
* learning
* goal
* project
* achievement
* interest
* gap

Confidence values:

* high
* medium
* low

JSON Schema:

{{
"date": "",
"person": "",
"memory_records": [
{{
"category": "",
"fact": "",
"status": "",
"confidence": ""
}}
]
}}

TEXT TO ANALYZE:

{knowledge}
"""

response = llm.invoke(prompt)

print("===== EXTRACTED MEMORIES =====")
print(response.content)


```
